```python
import pandas as pd
```

```python
# Re-align GPT2 columns
# Filter end/stat of sentence tokenz
# Then run the CogSci allignment script
```

```python
#Re-align gpt2 columns so that single tokens are on each column
def realign_gpt2(gpt2):
    gpt2 = gpt2.values.tolist()
    temp = []
    for tok in gpt2:
        if (tok[2][0] == "Ä ") or (tok[1] == 1):
            if tok[1] == 1:
                temp.append([tok[2], tok[3]])
            else:
                temp.append([tok[2][1:], tok[3]])

        else:
            temp[len(temp)-1][0] += tok[2]
            temp[len(temp)-1][1] += tok[3]
    return pd.DataFrame(temp, columns=["token", "surprisal"])
```

```python
ref = pd.read_csv("../data/sample_suites/test_items.csv", header=0, names=["test", "item", "condition", "region", "token"])
grnn = pd.read_csv("../data/sample_suites/model_output/grnn_output.txt", sep = "\t")[["token", "surprisal"]]
jrnn = pd.read_csv("../data/sample_suites/model_output/jrnn_output.txt", sep = "\t")[["token", "surprisal"]]
gpt2 = pd.read_csv("../data/sample_suites/model_output/gpt2_output.txt", sep = "\t")
gpt2 = realign_gpt2(gpt2)
ngram = pd.read_csv("../data/sample_suites/model_output/ngram_output.txt", sep = "\t")[["token", "surprisal"]]
```

```python
# Remove all of the eos tokenization
eos_toks = set(["<eos>", "</s>", "<S>", "</S>"])    

grnn = grnn[~grnn.token.isin(eos_toks)]
jrnn = jrnn[~jrnn.token.isin(eos_toks)]
gpt2 = gpt2[~gpt2.token.isin(eos_toks)]
ngram = ngram[~ngram.token.isin(eos_toks)]
```

```python
# Harmonize
print(len(ref))
print(len(grnn))
print(len(gpt2))
print(len(jrnn))
print(len(ngram))

print(ref.values.tolist()[0])
print(jrnn.values.tolist()[0])
```
